```python
# FinanceDataReader 주식 정보, 환율 정보
# CNY/KRW은 최근 것만 불러와짐
import FinanceDataReader as fdr
```

```python
usd_krw_df = fdr.DataReader('USD/KRW', '2020')
```

```python
usd_krw_df.columns
usd_krw_df.reset_index(inplace=True)
usd_krw_df.head()
```

```python
rates = ['USD/KRW', 'USD/JPY', 'CNY/KRW', 'JPY/KRW']
startdate = '2020'
```

```python
exchange_rate_df = pd.DataFrame()
exchange_rate_df['Date'] = usd_krw_df['Date']
exchange_rate_df.head()
```

```python
for rate in rates:
	column_name = rate
	print(column_name)

	temp = fdr.DataReader(rate, startdate)
	temp.reset_index(inplace=True)
	temp_df = pd.merge(temp[['Date', 'Close']], usd_krw_df[['Date']], on='Date', how='right')
	exchange_rate_df[column_name] = temp_df.Close
```

```python
exchange_rate_df.head()
```

```python
exchange_rate_df.set_index('Date', inplace=True)
```

```python
exchange_rate_df.tail()
```

```python
import numpy as np
import pandas as pd
import FinanceDataReader as fdr
```

```python
america_stock = ['DJI', 'IXIC', 'US500', 'RUTNU', 'VIX']
startdate = '2020'
```

```python
def csv_save(names, start_date):
	file_len = len(names)
	print('파일 개수 ', file_len)

	for name in names:
		try:
			df = fdr.DataReader(name, start_date)
			name_ = name.replace('/', '_')
			print(name_)
			df.to_csv('./data/' + name_ + '.csv', index=True)
			print('저장됨')
			print('=' * 100)
		except:
			print('정보 없음')
```

```python
csv_save(america_stock, startdate)
```

```python
csv_save(rates, startdate)
```

```python
kospi = ['KQ11', 'KS50', 'KS100', 'KRX100', 'KS200']
```

```python
csv_save(kospi, startdate)
```
