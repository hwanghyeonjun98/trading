```python
from tqdm import tqdm

modify_list = os.listdir('../data/20221121/min_fillrow/')
```

```python

for stock in tqdm(modify_list[567:]):
    temp_df = pd.read_csv(f'../data/20221121/min_fillrow/{stock}')

    if temp_df['시가'].isnull().sum() < 194571:
        temp_df1 = temp_df.dropna()

        volume = pd.DataFrame()
        for day in temp_df1.날짜.unique():
            temp = pd.DataFrame()
            temp['누적거래량'] = temp_df1[temp_df1['날짜'] == day][::-1]['거래량'].cumsum()
            volume = pd.concat([volume, temp[::-1]], axis=0)

        temp_df1 = pd.concat([temp_df1, volume], axis=1)
        temp_df1['누적체결매수수량'] = temp_df1['누적거래량'] - temp_df1['누적체결매도수량']
        temp_df1.drop(columns=['누적거래량','Unnamed: 0'], axis=1)
        colum_1 = temp_df1[temp_df1.columns[0:10]]
        colum_2 = temp_df1[temp_df1.columns[-1]]
        colum_3 = temp_df1[temp_df1.columns[10:-2]]

        temp_df2 = pd.concat([colum_1, colum_2, colum_3], axis=1)

        temp_df2.to_csv(f'../data/20221121/modify_fillrow/{stock}', encoding='utf-8 sig', index=False)
```

```python
min_fill_path = '../data/20221121/modify_fillrow/'
day_update_path = '../data/20221121/day_update/'
concat_path = '../data/20221121/concat/'

from tqdm import tqdm

def min_day_concat_info(min_fill_path = min_fill_path, day_update_path=day_update_path, concat_path= concat_path):
    min_fill_list = os.listdir(min_fill_path)
    day_update_list = os.listdir(day_update_path)
    min_fill_list = min_fill_list[:1200]
    day_update_list = day_update_list[:1200]
    for min_df, day_df in tqdm(zip(min_fill_list, day_update_list)):
        stock_min_df = pd.read_csv(f'{min_fill_path}{min_df}', encoding='utf-8 sig', index_col=0)
        stock_day_df = pd.read_csv(f'{day_update_path}{day_df}', encoding='utf-8 sig')
        
        input_list = ['전일대비','상장주식수','시가총액','외국인주문한도수량'
                    ,'외국인주문가능수량','외국인현보유수량','외국인현보유비율'
                    ,'수정주가일자','수정주가비율','기관순매수량','기관누적순매수량'
                    ,'등락주선','등락비율','예탁금','주식회전율','거래성립률']                                  
        list_ = ['시간','시가','고가','저가','종가','거래량','거래대금','누적체결매도수량'
                 ,'등락주선','등락비율','예탁금','주식회전율','거래성립률']
            
        stock_min_df = stock_min_df.drop(input_list, axis=1)
        stock_day_df = stock_day_df.drop(list_, axis=1)
        
        stock_day_df = stock_day_df.rename(columns={'Unnamed: 0':'날짜'})
        concat_df = pd.merge(stock_min_df, stock_day_df,on='날짜', how='left')
        
        concat_df.to_csv(f'{concat_path}concat_{min_df}', encoding='utf-8 sig', index=False)          
        
    return concat_df
```
